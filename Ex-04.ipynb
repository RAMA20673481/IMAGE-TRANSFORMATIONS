```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("Chennai.png")
```

```python
image.shape
```

```python
#Display the images.
plt.imshow(image[:,:,::-1])
plt.title("Original Image")
plt.show()
```

```python
# i) Image Translation
tx,ty=100,200
M_translation=np.float32([[1,0,tx],[0,1,ty]])
translated_image=cv2.warpAffine(image,M_translation, (673,419))
```

```python
plt.imshow(translated_image[:,:,::-1])
plt.title("Translated Image")
plt.axis("on")
plt.show()
```

```python
# Image Scaling
fx,fy=2.0,1.0
scaled_image=cv2.resize(image,None,fx=fx,fy=fy, interpolation=cv2.INTER_LINEAR)
```

```python
plt.imshow(scaled_image[:,:,::-1])
plt.title("Scaled Image")
plt.axis("on")
plt.show()
```

```python
# Image Shearing
shear_matrix=np.float32([[1,0.5,0],[0.5,1,0]])
sheared_image = cv2.warpAffine(image, shear_matrix, (673, 419))
```

```python
plt.imshow(sheared_image[:,:,::-1])
plt.title("Sheared Image")
plt.axis('on')
plt.show()
```

```python
# Image Reflection
reflected_image = cv2.flip(image, 2)
```

```python
# Show original image 
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image[:, :, ::-1])
plt.title("Original Image")
plt.axis('off')

# Show reflected image 
plt.subplot(1, 2, 2)
plt.imshow(reflected_image[:,:,::-1])
plt.title("Reflected Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Image Rotation
(height,width) = image.shape[:2]
angle = 45
center = (width // 2, height // 2)
M_rotation=cv2.getRotationMatrix2D(center,angle,1)
rotated_image = cv2.warpAffine(image, M_rotation, (width, height))
```

```python
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image")
plt.axis("off")
```

```python
# Image Cropping
x,y,w,h = 0,0,200,150
cropped_image = image[y:y+h, x:x+w]
```
